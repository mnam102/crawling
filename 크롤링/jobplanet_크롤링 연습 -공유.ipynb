```python
from selenium import webdriver
import pandas as pd

#풀스크린으로 바꾸는 옵션
from selenium.webdriver.chrome.options import Options
```

```python
options=Options()
options.add_argument('--start-fullscreen')

#페이지접속
driver =webdriver.Chrome('chromedriver',chrome_options=options)
driver.get('https://www.jobplanet.co.kr/companies/61420/reviews/%EB%B0%B0%EB%8B%AC%EC%9D%98%EB%AF%BC%EC%A1%B1')
```

```python
driver.find_element_by_xpath('//*[@id="premiumReviewChart"]/div/div[3]/div[2]/div/div[1]/button').click()
```

```python
driver.implicitly_wait(10)#페이지가 뜰때까지 기다려주는함수

#로그인
driver.find_element_by_xpath('/html/body/div[5]/div[1]/header/div/div[2]/a[1]/span').click()
```

```python
#이메일 및 패스워드 입력
driver.find_element_by_xpath('//*[@id="user_email"]').send_keys('개인아이디')#개인 아이디와 패스워드 입력.
driver.find_element_by_xpath('//*[@id="user_password"]').send_keys('개인패스워드')
```

```python
driver.find_element_by_xpath('//*[@id="signInSignInCon"]/div[2]/div/section[2]/fieldset/button').click()
```

```python
#직종전체로 
driver.find_element_by_xpath('//*[@id="occupation_select"]/option[1]').click()
```

```python
comm_list=[]
good_comm_list=[]
bad_comm_list=[]
category_list=[]

employment_list=[]
region_list=[]
date_list=[]
expectation_list=[]

grow_list=[]
recommend_list=[]
```

```python
for j in range(0,47):
    #데이터수집
    for i in range(1,6):
        try:
            comm=driver.find_element_by_xpath('//*[@id="viewReviewsList"]/div/div/div/section['+str(i)+']/div/div[2]/div/div[1]/h2').text
            good_comm=driver.find_element_by_xpath('//*[@id="viewReviewsList"]/div/div/div/section['+str(i)+']/div/div[2]/div/dl/dd[1]/span').text
            bad_comm=driver.find_element_by_xpath('//*[@id="viewReviewsList"]/div/div/div/section['+str(i)+']/div/div[2]/div/dl/dd[2]/span').text
            category=driver.find_element_by_xpath('//*[@id="viewReviewsList"]/div/div/div/section['+str(i)+']/div/div[1]/span[2]').text

            employment=driver.find_element_by_xpath('//*[@id="viewReviewsList"]/div/div/div/section['+str(i)+']/div/div[1]/span[4]').text
            region=driver.find_element_by_xpath('//*[@id="viewReviewsList"]/div/div/div/section['+str(i)+']/div/div[1]/span[6]').text
            date=driver.find_element_by_xpath('//*[@id="viewReviewsList"]/div/div/div/section['+str(i)+']/div/div[1]/span[8]').text
            expectation=driver.find_element_by_xpath('//*[@id="viewReviewsList"]/div/div/div/section['+str(i)+']/div/div[2]/div/dl/dd[3]/span').text
            grow=driver.find_element_by_xpath('//*[@id="viewReviewsList"]/div/div/div/section['+str(i)+']/div/div[2]/div/p[1]/strong').text
            recommend=driver.find_element_by_xpath('//*[@id="viewReviewsList"]/div/div/div/section['+str(i)+']/div/div[2]/div/p[2]/em').text

            comm_list.append(comm)
            good_comm_list.append(good_comm)
            bad_comm_list.append(bad_comm)
            category_list.append(category)

            employment_list.append(employment)
            region_list.append(region)
            date_list.append(date)
            expectation_list.append(expectation)
            grow_list.append(grow)
            recommend_list.append(recommend)
        except Exception as e:
            print(e) #pass
        
        finally:
            pass

    driver.find_element_by_xpath('//*[@id="viewReviewsList"]/div/div/div/div[2]/article/a[7]').click()
```

```python
data1=pd.DataFrame({'직종':category_list, '재직여부':employment_list, '지역':region_list,'작성시간':date_list,
            '한줄평':comm_list,'기업장점':good_comm_list,'기업단점':bad_comm_list,'바라는점':expectation_list,'성장여부':grow_list,
              '추천여부':recommend_list
             })
print(data1.shape)
```
